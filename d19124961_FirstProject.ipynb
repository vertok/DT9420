```python
# TU DUBLIN
# Alexey Obukhov D19124961@mytudublin.ie Erasmus+ student from Germany, Dresden
# programm the Hangman Game: it reads the file words.txt which is in the same ordner as this file,
# gets words list from there for the game and chooses one for the game. user have n*length of the 
# word attemps to find the word. programm will make recods with winners in log.txt 
# to start the program just press "Run". Good luck!

# functions

# this function returns random word from words.txt file which
# should be situated in the same folder as this file
def get_the_word():
    import random
    f=open('words.txt') # file words.txt should be in the same folder as this file
    lines = f.readlines()
    return str(lines[random.randint(0,48)])
    f.close()
    
# this function is responcible for proving whether the letter 
# is in the word or not
def find_letter(letter):
    prov = 0 # for output the whole word during wrong try
    for pos, char in enumerate(text):
        if char == letter:
            user_progress[pos]=letter
            user_progress_formatted = ' '.join(user_progress)
            prov = 1
            print(user_progress_formatted)
    if(prov == 0):
            user_progress_formatted = ' '.join(user_progress)
            print(user_progress_formatted)
            
# this funktion proves whether the letter is in the word
def score_collection(letter):
    global i # number of attemps from the main body
    global score_collect # number of users score
    if(len(letter)==1):
        if(letter.isalpha()):
            # this part proved whether the letter was already chosen by user
            if (any(letter in word for word in text.split())):
                if(letter in right_letters):
                    print("you have tried already with this letter")
                    i -= 1 # number of attemps decreases but score_collect is the same
                else:
                    right_letters.append(letter)
                    right_letters.sort()
                    score_collect += 20
                print("right letters: ",right_letters,"\nwrong letters: ",wrong_letters) 
            else:
                if(letter in wrong_letters):
                    print("you have tried already with this letter")
                    i -= 1
                else:
                    wrong_letters.append(letter)
                    wrong_letters.sort()
                    score_collect -= 1
                print("right letters: ",right_letters,"\nwrong letters: ",wrong_letters)
            print("positions:",[pos+1 for pos, char in enumerate(text) if char == letter]) 
        else:
            score_collect -= 1
    else:
        score_collect -= 1
        
# this function makes output whether the word was found
def print_into_file():
    global i
    print("congratulation! you've finished the game using",i,"attends for finding the word:",text,"well done!")
    print("type please your name for the records:")
    name = input()
    log = open('log.txt', 'a')
    temp = "\nword: " + str(text) + "number of tries: " + str(i) + "\nplayer: " + str(name) +"\nscore: " + str(score_collect) + "\n"
    log.write(temp)
    log.close()
    
    
# main body of the program
import sys
import os

text = get_the_word().lower() # assigning the word from *.txt to 'text'
num_letters = len(str(text)) - 1
wrong_letters = []
right_letters = []
score_collect = 0
k = 0 # this counter k counts the number of found letters in the word
i = 1 # number of current attemp
n = 2*num_letters #number of attemps is 2*length of word
tries = 0 # number of attemps

print("############ WELCOME! ##############\n####### TO THE HANGMAN GAME ########\nYOU HAVE",
        n,"ATTEMPS TO FIND THE WORD \nHIDDEN BELLOW\n")
user_progress = ["_" for x in range(num_letters)]

user_progress_formatted = ' '.join(user_progress)
# to avoid output like ['_', '...] because user_progress is a list
# it will be converted to the string with this line of code

print(user_progress_formatted.upper())
#print(text) # prints the word. just for testing

# main body is while-loop which will be escaped whether the word was found or there will be no more attemps or user writes q!
while(i <= n):
    print("to quit the game anytime and show the word write q!")
    if(n - i + 1 != 1):
        print("\nyou have",n-i+1,"attemps left")
    else:
        print("\nyou have your last attemp")
    print("enter the letter: ")
    tries += 1
    letter = input().lower()

    # this part will calculate the number of found letters
    for char in text:
        if(str(char) == letter):
            if str(char) not in str(right_letters):
                k += 1

    score_collection(letter)
    find_letter(letter) 


    # exit from program
    print("current score:",score_collect)
    if(letter == "q!"):
        print("good buy! the word was:", text)
        break
    if(k == len(text)-1): # user wins here
        print_into_file()
        break
    if(i == n): # user loses here
        print("you've lost this round because you don't have any more attemps. The word was:",text)
        break        
    i += 1
# the end.
```
